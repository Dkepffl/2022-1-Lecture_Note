#### 2. 순서도 실습

```python
def average(x):
  i = 0
  sum = 0
  while i < len(x):
    sum += x[i]
    i += 1
  return sum / len(x)

x = [1,2,3,4,5]
y = average(x)
print(y)
```

```python
def average(x):
  sum = 0
  for i in range(len(x)):
    sum += x[i]
  return sum / len(x)

x = [1,2,3,4,5]
y = average(x)
print(y)
```

```python
def func(x):
  a = []
  for i in range(1, x + 1):
    if x % i == 0:
      a.append(i)
  return a
  
x = 12
y = func(x)
print(y)
```

#### | **Example #4**
- 목표 : 사용자에게 양의 정수 n을 입력 받아, 피보나치 수열의 n번째 값을 출력하는 프로그램을 작성하시오.
- 문제 해결 알고리즘 : 피보나치 수열
- 필요한 변수 : `var0`, `var1`, `var2`
- 함수 정의

```python
def fibonachi():
  n = int(input())
  x = [1,1] 
  for i in range(0, n):
    x.append(x[i]+x[i+1])
  print(x)
  return x[n-1]
```

```python
fibonachi()
```

```python
# Solution
n = int(input())

var1 = 1
var2 = 0
i = 2

while i <= n:
  var0 = var1 + var2
  var2 = var1
  var1 = var0
  i += 1
print(var0)
```

#### | **Example #1**
- 목표 : 주어진 수 나열
- 문제 해결 알고리즘 : 숫자 비교
- 함수 이름, 입출력 설정 : `Max()`, `return`
- 함수 정의

```python
def Max(a, b, c):
  if a > b:
    if a > c:
      return a
    else:
      return c
  else:
    if b > c:
      return b
    else:
      return c
```

```python
Max(4, 20, 1)
```
